```python
import httpx
import asyncio
import pandas as pd
from IPython.display import clear_output

async def send_request_without_waiting(request_body:dict):
    async with httpx.AsyncClient() as client:
        # –û—Ç–ø—Ä–∞–≤–∏—Ç—å –∑–∞–ø—Ä–æ—Å –±–µ–∑ –æ–∂–∏–¥–∞–Ω–∏—è –æ—Ç–≤–µ—Ç–∞
        print("request send")
        response = await client.post('http://0.0.0.0:8090/', json=request_body, timeout=60*3)
        response_data = response.json()
        status_code = response.status_code
        return status_code, response_data
```

```python
import asyncio
import random
import pandas as pd
import time
from IPython.display import clear_output

async def main(errors):
    columns = ['Episode', 'Video', 'Status_Code', 'TimeSleep', 'ResponseTime', 'ServerLoads', 'MRT', 'QL',]
    df = pd.DataFrame(columns=columns)
    for idx, request_body in enumerate(dataset):
        try:
            start_time = time.time()
            task = asyncio.create_task(send_request_without_waiting(request_body))
            status_code, response_data = await task
            response_time = time.time() - start_time
            sl = response_data.get("server_loads")
            mrt = response_data.get("mrt")
            ql = response_data.get("queue_len")
            time_sleep = random.randint(0, 3)
            row = {
                'Episode': idx,
                'Video': request_body['video'],
                'Status_Code': status_code,
                'TimeSleep': time_sleep,
                'ResponseTime': response_time,
                'ServerLoads': sl,
                'MRT': mrt,
                'QL': ql,
                'CountErrors': len(errors),
            }
            df.loc[len(df)] = row

            if (idx + 1) % 10 == 0:
                clear_output()
                print(df.to_string(index=False))

            await asyncio.sleep(time_sleep)

        except httpx.ReadTimeout as e:
            errors.append(idx)
        except httpx.ConnectError as e:
            print("server was destroyed")
            return df
        except httpx.RemoteProtocolError as e:
            print("server was destroyed")
            return df

    return df
```
